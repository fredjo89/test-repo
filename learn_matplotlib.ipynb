```python
import matplotlib
print(matplotlib.__version__)

import matplotlib.pyplot as plt
import numpy as np
```

# Markers

```python
# Using the format string ftm: marker|line|color
# Keyword argument: Marker size (ms)
# Keyword argument: Marker Edge Color (mec)
# Keyword argument: Marker Face Color (mfc)
x = np.array([1,2,6,8])
y = np.array([3,8,1,10])

plt.plot(x,y, 's--r', ms = 20, mec = 'b', mfc = '#4CAF50')
plt.show()
```

# Line

```python
# Keyword Argument: linestyle (ls)
# Keyword Argument: color (c)
# Keyword argument Line Width (lw)
y1 = [3,8,1,10]
y2 = [7,8,3,3,12]
plt.plot(y1, ls = 'dotted', c = 'k', lw = 5) # ls = ':' would do the same
plt.plot(y2, ls = 'solid', c = 'b', lw = 1) 

# xlabel, ylabel, title, font-dict
# Keyword Argument: location (loc)
title_font_dict = {'family': 'serif', 'color': 'pink', 'size': '20'}
plt.xlabel('my x-label')
plt.ylabel('my y-label')
plt.title('My Figure Title', title_font_dict, loc = 'center')

plt.show()
```
